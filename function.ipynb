```python
def add(a,b):
    print(a+b)
def sub(c,d):
    print(c-d)
def mul(e,f):
    print(e*f)
def div(a,b):
    print(a/b)
a = int(input("Enter a = "))
b = int(input("Enter b = "))
o = input("Enter + - * /  ")
if o == "+":
    add(a,b)
elif o == "-":
    sub(a,b)
elif o == "*":
    mul(a,b)
elif o == "/" and b !=0:
    div(a,b)
elif o == "/" and b ==0:
    print("The value of b can not be zero")
else:
    print("Invalid Operator")
```
